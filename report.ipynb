## Dataset

The dataset consists of GPAs in multiple courses at **UIUC** taught by multiple instructors. It is taken from the [source](https://github.com/wadefagen/datasets/blob/master/gpa/uiuc-gpa-dataset.csv). A typical row in this dataset consists of multiple attributes. Some important attributes are explained below.

- *__Year__* - Year in which the course was taught, for example, 2018, 2019 etc.
- *__Term__* - Semester in which the course was taught, such as, Spring, Fall etc.
- *__YearTerm__* - Combination of Year and Term, such as, 2018-sp.
- *__Subject__* - Department offering the course such as STAT, CS etc.
- *__Number__* - Course number
- *__Course Title__* - Title of the course
- *__A+__* - Number of students who got this grade
- *__A__* - Number of students who got this grade
- *__A-__* - Number of students who got this grade
- *__B+__* - Number of students who got this grade
- *__B__* - Number of students who got this grade
- *__B-__* - Number of students who got this grade
- *__C+__* - Number of students who got this grade
- *__C__* - Number of students who got this grade
- *__C-__* - Number of students who got this grade
- *__D+__* - Number of students who got this grade
- *__D__* - Number of students who got this grade
- *__D-__* - Number of students who got this grade
- *__F__* - Number of students who got this grade
- *__W__* - Number of students who got this grade
- *__Primary Instructor__* - Name of the instructor teaching the course

## Question of Interest

*The aim of this report is to create a dynamic graph to demonstrate the grade distributions for each course that a particular instructor has taught of a particular subject(department)*.

A web-based application is developed using __Dash__ framework and graphs are plotted using the library __plotly__ for interactivity such hover, zoom etc.

## Testing of the web app

For example, if the subject is changed from CS to Stats, then only those instructors are shown who are part of the STATS department

- For the demonstration, the __STATS__ department is chosen as the __Subject__ and __Prof. Andrew Culpepper__ is chosen as the __Instructor__.

- And then the __Subject__ is changed to __CS__ with a random instructor

## Conclusion & Discussion

From images in the above section, it can be clearly viewed that when the choice of each __Subject__ and __Instructor__ changes, the graph also changes. It shows the distribution of grades in all the courses that the particular instructor has taught over multiple years and terms. This distribution is shown in terms of a histogram where each bar represents the grade distribution amongst different courses which are color coded. In short, the overall graph represents the distribution amongst different grades for different courses. This web-application categorically achieves the objective set in the previous sections.

For improvements, there are few things that could be done. Different flavours of histograms such as horizontal style could be added to improve interpretability. Morever, experimenting with other plots such as scatter plot could be used to check for correlations in the data.
